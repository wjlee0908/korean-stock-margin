```python
import pandas as pd                
import matplotlib as mpl          
import matplotlib.pyplot as plt    

# csv 파일을 읽어서 DataFrame 객체로 만듦. 인덱스 컬럼은 종목코드로 설정
corporation_information_filename = 'corporation_information.csv'
df = pd.read_csv(corporation_information_filename, index_col='종목코드', encoding='utf-8')


# Windows 한글 폰트 설정
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
```

```python
df = df[["회사명","기업가치","시가총액","안전마진율"]]
df = df.sort_values(by='안전마진율', ascending=False)
df
```

```python
# 안전마진율 그래프
margin_df = df[["회사명","안전마진율"]]
margin_df = margin_df.set_index("회사명")

ax = margin_df.plot(kind='bar', figsize=(20, 10), title='기업별 안전마진율', legend=True, fontsize=1)
ax.legend(['안전마진율(%)'], fontsize=12)    # 범례 지정
```

```python
# 기업가치 데이터프레임
value_df = df[["회사명", "기업가치", "시가총액"]]
value_df = value_df.set_index("회사명")
value_df = value_df.sort_values(by='기업가치', ascending=False)
value_df
```

```python
# 기업가치 상위 50개 데이터프레임
high_value_df = value_df.iloc[:50]
high_value_df
```

```python
# 기업가치 순위 그래프

ax = high_value_df.plot(kind='bar', figsize=(20, 10), title='기업별 가치 - 기업가치순', legend=True, fontsize=12)
ax.legend(['기업가치(원)', '시가총액(원)'], fontsize=12)    # 범례 지정
```

```python
# 시가총액 순위 그래프
high_price_df = value_df.sort_values(by='시가총액', ascending=False).iloc[:50]
ax = high_price_df.plot(kind='bar', figsize=(20, 10), title='기업별 가치-시가총액순', legend=True, fontsize=12)
ax.legend(['기업가치(원)', '시가총액(원)'], fontsize=12)    # 범례 지정
```

```python
# 안전마진 비율 그래프
high_margin_df = margin_df.iloc[:100]
ax = high_margin_df.plot(kind='bar', figsize=(20, 10), title='기업별 안전마진율', legend=True, fontsize=12)
ax.legend(['안전마진율(%)'], fontsize=12)    # 범례 지정
```

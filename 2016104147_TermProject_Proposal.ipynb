## 1. 주제 선정 이유

주식 투자 방법 중 가치투자는 기업의 실적과 재산으로 가치를 분석하여 실제 가치보다 저평가된 기업에 투자하는 방법이다. 워런 버핏, 벤자민 그레이엄, 피터 린치는 가치 투자 방법으로 높은 수익률을 기록하여 유명해졌다. 이들은 책을 출판하여 사람들에게 가치 투자 방법을 추천한다. 하지만 한국의 개인 투자자들 사이에서 가치 투자 방법은 환영받지 못 한다. 단기 수익률이 낮고, 한국 주식의 주가가 가치에 따라 움직이지 않는다고 생각하는 사람이 많기 때문이다. <br>
나 또한 가치 투자에 관한 책을 읽고 관심 있는 기업을 정성·정량적인 방법으로 분석하고 투자를 했었다. 정성적인 방법으로는 기업이 하는 사업을 분석하고 관련 신문 기사를 보면서 사업의 안정성, 유망성을 판단하였고 정량적인 방법으로는 재무제표를 확인하여 기업의 수익, 부채, 재산을 분석했다. 또, 『현명한 초보 투자자』라는 책에서 재무제표에 표현된 수치로 기업의 가치를 계산하는 공식이 나와 있었는데, 이 공식을 스프레드시트로 계산해 안전 마진을 구하고 이를 활용했었다. 그러나 1년 동안 투자한 결과, 수익률 -15%로 손해를 봤다. <br>

그래서 웹파이선 프로그래밍 강좌를 수강하며 배운 지식을 활용하여 재무제표가 공개된 기업들의 재무제표상의 가치와 과거의 데이터를 통하여 안전 마진(기업 가치와 주가의 갭)과 주식 수익률 사이에 상관관계가 실제로 존재하는지 확인해 보기로 했다.

## 2. 가설 정의

1) 안전 마진(기업 가치와 시가 총액의 차이 비율. 기업가치-시가총액/시가총액)과 수익률 간의 상관 관계가 있는 지 관찰한다. 과거 특정 시점의 기업의 안전 마진을 구하고 안전 마진의 크기 순서에 따라 그룹화한다. 그룹별로 1년 후, 5년 후, 10년 후의 평균 수익률을 구한다. 시장 수익률(KOSPI, KOSDAQ 전체 수익률)도 표시하여 비교한다. <br>
2) 안전 마진 상위 50개 기업 개별로 수익률 분포를 확인한다. 평균의 함정에 빠지지 않았는지 검증한다. <br>
3) 만약 안전 마진과 수익률 사이에 유의미한 상관 관계가 없다면, PER, PBR과 같은 가치 판단 지표를 기준으로도 수익률과의 상관 관계를 확인한다. <br>
4) 현재 기업 가치를 기준으로 안전 마진을 계산하여 안전 마진 상위 기업을 표시한다. <br>
- 재무상태표를 표시할 수 없는 종목(ETF, SPAC주 등)은 조사 대상에서 제외한다. <br>
- 조사 기간 내 재무상태표가 존재하지 않는 신생 기업도 조사 대상에서 제외한다. <br>
- n년 후 상장 폐지된 기업의 주가는 0으로 놓는다. 즉 수익률은 -100%이다. <br>

『현명한 초보 투자자』에 따르면, 기업의 가치를 구하는 방법은 다음과 같다.<br>
**기업의 가치 = 사업 가치 + 재산 가치 - 부채<br>**

**사업 가치 = (영업이익 - 세금) / 기대수익률<br>**
한국의 법인세 실효세율은 2017년 기준으로 과표 구간 5천억원 초과 기업은 18.0%, 표 구간 1천억~5천억원 기업은 20.5%,  500억~1천억원 기업들은 19.5%, 200억~500억원 기업들은 19.0%이다. 사업 가치를 계산할 때는 약 20.0%로 세율을 계산한다. 이 때, 한 해의 영업이익에서 세금을 뺀 값은 영업이익 - 영업이익 * 0.2이 된다. <br>
주식의 기대수익률은 대체로 5~9%이나 회사채 금리는 3년간 1.6%~2.6%이므로 대략적인 중간 지점인 6%를 투자자와 은행 모두가 기대하는 수익률로 놓고 계산한다.<br>
따라서 사업가치를 계산하는 식은 다음과 같다.<br>
사업가치 = (영업이익 - 영업이익 * 0.2) / 0.06 = **영업이익 * 13.3 <br>**

__재산 가치 = 유동자산 - 상장기업의 유동비율 * 유동부채 + 투자자산__<br>
한국의 상장사 유동비율 평균은 129%로 1.29로 놓고 계산한다.<br>
기업의 투자자산은 재무제표 상의 장기금융상품, 매도가능증권, 만기보유증권, 지분법적용 투자주식, 장기대여금, 투자부동산, 장기투자자산을 합한 값으로 계산한다.<br>
따라서 재산가치를 계산하는 식은 다음과 같다.<br>
__재산가치 = 유동자산 - 1.29 * 유동부채 + 투자자산__ <br>
__투자자산 = 장기금융상품 + 매도가능증권 + 만기보유증권 + 지분법적용 투자주식 + 장기대여금 + 투자부동산 + 장기투자자산__ <br>

부채는 고정부채만 반영한다.

위의 내용을 정리하면 다음과 같다.<br>
__기업 가치 = 사업 가치 + 재산 가치 - 부채__<br>
__=(영업이익 * 13.3) + (유동자산 - 1.29 * 유동부채 + 투자자산) - 고정부채__

## 3. 인터넷을 통한 데이터 획득 방법

기업의 평가된 가치와 주식의 수익률 2가지 정보를 비교해야 하므로, 기업의 재무제표 데이터와 과거 주가 기록 데이터를 획득해야 한다.<br>

재무제표 데이터는 NICE평가정보의 상장온라인 서비스에서 재무제표를 BeautifulSoup 모듈을 이용해 크롤링하여 얻는다.<br>

<br>
기업 가치를 평가하기 위해 표에서 다음 행을 획득하여야 한다.<br>
유동자산(계), 유동부채(계), 장기투자자산, 비유동부채(계) <br>
영업이익은 재무상태표 아래의 현금흐름표의 영업이익(손실) 행을 통해 획득할 수 있다. <br>

`requests` 모듈로 상장온라인 url에서 html 파일을 전송받고, Beautifulsoup 모듈로 태그 분석을 하였다. 연간 재무제표는 `<div name="yt1">`이므로, name이 yt1인 테이블을 찾아 정보를 탐색할 타겟으로 정했다. 먼저, 연도별로 구성된 열에서 조사할 연도의 열 인덱스를 `self.__find_year_column_index()` 메소드에서 반환한다.BeautifulSoup 모듈의 `.find()` 함수를 이용해 테이블의 첫번째 <tr>(table row) 태그를 찾고, 그 안에 있는 <th>(table header) 태그 중에서 조사할 연도 문자가 몇 번째 열에 있는지 찾아 그 인덱스를 반환한다. 그 다음에는 `self.__generate_dict_from_html_table()` 메소드를 통해 header 밑의 tr과 td데이터를 분석하여 유동자산, 유동부채, 투자자산, 고정부채 정보를 얻었다.

과거 시가총액 데이터는 NICE평가정보의 상장온라인 서비스에서 주식 및 주식수 표를 BeautifulSoup 모듈을 이용해 크롤링하여 얻는다.

## 6. 결론

__시가총액이 높은 모든 기업이 기업 가치가 높지는 않았다.__<br>
재무 정보를 바탕으로 평가한 기업 가치 순위는 시가총액 순위와는 차이가 있었다.<br>
[그래프 1]과 [그래프 2]를 비교했을 때, 삼성전자, 에스케이하이닉스, LG화학 등의 기업은 기업 가치와 시가 총액 모두 상위 50개 이내였으나, 현대자동차, lg디스플레이, 한국전력공사 등의 기업은 시가총액의 순위는 상위 50개 이내이지만, 기업 가치 순위는 그렇지 않았다. 특히, 한국전력공사, 현대중공업, lg디스플레이는 시가 총액이 높지만, 기업 가치는 음의 값을 가졌다.<br>
<br>
__상장 기업의 주식은 고평가된 기업이 많았다__<br>
[그래프 3]은 상장 기업의 안전마진율 분포를 보여준다. 안전마진율이 양수인 기업보다 음수인 기업이 더 많았다. 즉, 기업 가치보다 시가총액이 더 고가인 기업이 많아 시장 가치가 고평가되어 있음을 확인하였다.<br>
안전마진율이 가장 높은 기업은 반도체 제조 장비를 만드는 피에스케이이다. 안전마진 상위 100개 기업 중 인지도 있는 기업은 한신공영, 아디다스홀딩스, 신한, 에스케이하이닉스, GS, 두산인프라코어, 한화 등이 있었다.

## 10. 별첨 3 - 데이터 변환 작업 등을 위해서 개발한 Python 소스 코드

__company.py__ : 기업의 재무 정보를 크롤링하고, 기업 가치를 계산하는 모듈

```python
import requests
from bs4 import BeautifulSoup
import currency

class Company:
    def __init__(self, stock_code, year):
        self.__stock_code = stock_code
        self.__year = year
        self.__name = ''    # 기업명
        self.__asset = 0    # 유동자산
        self.__current_liabilities = 0    # 유동부채
        self.__fixed_liabilities = 0    # 고정부채
        self.__investment_asset = 0    # 투자자산
        self.__profit = 0    # 영업이익
        self.__value = 0    # 기업가치
        
        self.__is_scraping_succeed = self.__set_information_from_table(self.__stock_code)

        if self.__is_scraping_succeed:
            self.__value = self.calculate_value()

    def __str__(self):
        return '기업명: {}  종목코드: {}    유동자산: {}    유동부채: {}    고정부채: {}    투자자산: {}    영업이익: {}    기업가치: {}'\
            .format(self.__name, self.__stock_code, self.__asset, self.__current_liabilities, self.__fixed_liabilities, \
                self.__investment_asset, self.__profit, self.__value)

    def get_information_list(self):
        return [self.__stock_code, self.__name, self.__asset, self.__current_liabilities, \
        self.__fixed_liabilities, self.__investment_asset, self.__profit, self.__value]

    def get_value(self):
        return self.__value
    
    def is_valid(self):
        return self.__is_scraping_succeed

    def calculate_value(self):
        business_value = self.__profit * 13.3
        asset_value = self.__asset - (1.29 * self.__current_liabilities) + self.__investment_asset
        return business_value + asset_value - self.__fixed_liabilities

    def __set_information_from_table(self, stock_code):
        # 재무제표 있는 기업정보 사이트(상장온라인)의 주소
        company_information_url = "http://media.kisline.com/fininfo/mainFininfo.nice"
        table_page_num = 4    # 재무제표 있는 페이지 번호. nav 파리미터

        # 상장온라인 사이트에서 재무정보를 받아 멤버로 설정.
        money_unit = 1000000    # 표에서 돈의 기본 단위

        response = requests.get(company_information_url, {'paper_stock': stock_code, 'nav':str(table_page_num)})

        # 페이지 읽어오기 실패하면 종료
        if response.status_code != 200:
            print("페이지가 응답하지 않습니다.")
            return False

        # html 파일을 beatutifulsoup로 파싱
        soup = BeautifulSoup(response.content, 'html.parser')
        if type(soup) == type(None):
            return False

        # 회사명 탐색
        info_div = soup.find('div', {'class':'cot'})
        if type(info_div) == type(None):
            return False
        name_tag = info_div.find('strong')
        if type(name_tag) == type(None):
            return False
        
        self.__name = name_tag.text

        # 재무정보 테이블 탐색
        row_to_search = ['유동자산(계)', '유동부채(계)', '장기투자자산', '비유동부채(계)']
        financial_table = soup.find('div', {'name':'yt1'})
        year_index = self.__find_year_column_index(financial_table, self.__year)
        if year_index == None:
            return False
        financial_dict = self.__generate_dict_from_html_table(financial_table, row_to_search, year_index)

        # 테이블에서 재무정보가 찾아야 하는 행의 갯수와 다르면 종료.
        if len(financial_dict) != len(row_to_search):
            return False
            
        for key, value in financial_dict.items():
            # 상장온라인 표에서 0을 '-'으로 표기해서 변환해줌
            if financial_dict[key] == '-':
                financial_dict[key] = '0'
            financial_dict[key] = currency.convert_to_int(financial_dict[key], money_unit)

        self.__asset = financial_dict['유동자산(계)']
        self.__current_liabilities = financial_dict['유동부채(계)']
        self.__investment_asset = financial_dict['장기투자자산']
        self.__fixed_liabilities = financial_dict['비유동부채(계)']

        # 포괄손익계산서 테이블 탐색
        row_to_search = ['영업이익(손실)']
        profit_table = soup.find('div', {'name':'ypt1'})
        year_index = self.__find_year_column_index(financial_table, self.__year)
        if year_index == None:
            return False
        profit_dict = self.__generate_dict_from_html_table(profit_table, row_to_search, year_index)

        for key, value in profit_dict.items():
            # 상장온라인 표에서 0을 '-'으로 표기해서 변환해줌
            if profit_dict[key] == '-':
                profit_dict[key] = '0'
            profit_dict[key] = currency.convert_to_int(profit_dict[key], money_unit)
        
        self.__profit = profit_dict['영업이익(손실)']

        return True

    def __generate_dict_from_html_table(self, table_html, row_headers, column_index):
        """ (bs4.element.Tag, list, int) -> dict
        
        read data from html tag form table and store it in dictionary.
        And returns dictionary.
        
        financial_table = financial_table.find('div', {'name':'yt1'})
        >>> generate_dict_from_html_table(financial_table, ['profit', 'debt'], 4)
        {'profit' : '300,000', 'debt' : '20,000'}
        """

        table_info = dict()

        for tr in table_html.find_all('tr'):
            header_name = tr.find('th')

            # 찾을 행 있을 때만 진행
            if type(header_name) == type(None):
                continue

            # 찾을 행과 헤더가 동일한 행의 값만 읽음
            if header_name.text in row_headers:
                tds = list(tr.find_all('td'))    # 행의 모든 열 정보 list로 저장
                table_info[header_name.text] = tds[column_index].text

        return table_info

    def __find_year_column_index(self, table_tag, year):
       head_row = table_tag.find('tr')    # 테이블에서 첫 번째 row
       headers = head_row.find_all('th')

       for idx, th in enumerate(headers):
           header_name = th.text
           if str(year) + '.' in header_name:
               return idx - 1    # data 첫 열은 th라서 data index는 -1

       return None
```

__currency.py__: 표의 000,000,000 또는 - 형태로 표현되어 있는 금액 정보를 int형으로 변환해 주는 모듈

```python
def convert_to_int(currency_str, unit=1):
    """ (str) -> int
    
    convert string type currency value to integer value.
    it removes comma and multiplies its omitted unit.
    
    >>> convert_to_int('123,456', 100)
    12345600
    """
    currency_int = int(currency_str.replace(',','')) * unit
    return currency_int
```

__StockValue.py : 모든 기업을 객체로 생성하고, 기업 정보를 csv 파일로 출력하는 메인 모듈__

```python
from multiprocessing import Process, Manager
import random
import time
start_time = time.time() 

from company import *
import korean_market
import csv

def add_new_company(company_list, codes):
        for code in codes:
            new_company = Company(code, 2018)
            if new_company.is_valid() == True:
                company_list.append(new_company)
            else:
                del new_company    # 크롤링 실패한 데이터는 삭제


if __name__ == "__main__":

    with Manager() as manager:
        companies = manager.list()
        processes = []
        num_processes = 4
        stock_codes = korean_market.all_stock_codes
        #stock_codes = random.sample(korean_market.all_stock_codes, k=100)    # [TEST] 테스트용 샘플 100개 추출
        len_single_list = len(stock_codes) // num_processes

        # 리스트 하나를 process 개수만큼 나눠서 병렬 처리
        for i in range(num_processes):
            idx_start = i * len_single_list
            if i == num_processes - 1:
                idx_end = len(stock_codes)
            else:    
                idx_end = len_single_list * (i+1)
            codes = stock_codes[idx_start:idx_end]

            # 모든 상장 기업 정보 크롤링하는 과정 멀티프로세싱
            p = Process(target=add_new_company, args=(companies, codes))
            p.start()
            processes.append(p)
        for p in processes:
            p.join()
        
        with open('corporation_information.csv', 'w', encoding='utf8', newline='') as file_write:
            csv_writer = csv.writer(file_write)
            csv_writer.writerow(['종목코드', '회사명', '유동자산', '유동부채', '투자자산', '비유동부채', '영업이익', '기업가치'])

            for com in companies:
                    csv_writer.writerow(com.get_information_list())
        

        
        print("--- {} seconds ---".format(time.time() - start_time))
```
